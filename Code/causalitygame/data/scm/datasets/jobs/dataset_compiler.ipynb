```python

import pandas as pd
```

```python
columns = ["treatment", "age", "education", "black", "hispanic", "married", "nodegree", "RE75", "RE78"]

df_treated = pd.read_csv("nsw_treated.txt", delimiter="  ", names=columns)
df_control = pd.read_csv("nsw_control.txt", delimiter="  ", names=columns)
```

```python
df_merged = pd.concat([df_treated, df_control]).astype({
    "treatment": bool,
    "age": int,
    "education": int,
    "black": bool,
    "hispanic": bool,
    "married": bool,
    "nodegree": bool
})
df_merged
```

```python
df_merged.rename(columns={
    c: f"t:{c}" for c in ["treatment"]
}).rename(columns={
    c: f"c:{c}" for c in df_merged.columns if c not in ["treatment", "RE78"]
}).rename(columns={
    c: f"o:{c}" for c in ["RE78"]
}).to_csv("nsw.csv", index=False)
```
